# DATA PREPROCESSING

Creating a function to extract numeric numbers

```python
import re

def extract_numeric_average(text_value):
    # 1. If the row is empty (NaN), return None right away
    if pd.isna(text_value):
        return None
    
    try:
        # Convert to string just in case it's mixed
        text_str = str(text_value)
        
        # 2. Find all numbers (including decimals) in the text row
        # This will find ['10', '20'] out of '10-20 kg'
        numbers = re.findall(r'[-+]?\d*\.\d+|\d+', text_str)
        
        # 3. If we didn't find any numbers, return None
        if not numbers:
            return None
        
        # 4. Convert our found string numbers into actual float numbers
        float_numbers = [float(n) for n in numbers]
        
        # 5. Calculate the average of the found numbers (handles single numbers AND ranges!)
        return sum(float_numbers) / len(float_numbers)
        
    except Exception:
        # If anything weird happens, don't crash, just skip the row
        return None
```

```python
# Clean the columns using your new function
df['Clean_Lifespan'] = df['Lifespan (years)'].apply(extract_numeric_average)
df['Clean_Gestation'] = df['Gestation Period (days)'].apply(extract_numeric_average)
df['Clean_Offspring'] = df['Offspring per Birth'].apply(extract_numeric_average)

# Check the first 5 rows to see the old columns next to your beautiful new clean ones!
df[['Lifespan (years)', 'Clean_Lifespan', 'Offspring per Birth', 'Clean_Offspring']].head()
```

```python
df['Clean_Height'] = df['Height (cm)'].apply(extract_numeric_average)
df['Clean_Weight'] = df['Weight (kg)'].apply(extract_numeric_average)
df['Clean_AvgSpeed'] = df['Average Speed (km/h)'].apply(extract_numeric_average)
df['Clean_TopSpeed'] = df['Top Speed (km/h)'].apply(extract_numeric_average)

# Let's run df.info() one more time to see our beautiful new numeric columns added!
df.info()
```
