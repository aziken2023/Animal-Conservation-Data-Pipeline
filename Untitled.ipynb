```python
#Import Libraries
import pandas as pd
import seaborn as sns 

#Read and inspect file
df = pd.read_csv("Animal Dataset.csv")
df.info()
df.head()
```

# DATA PREPROCESSING

Creating a function to extract numeric numbers

```python
import re

def extract_numeric_average(text_value):
    # 1. If the row is empty (NaN), return None right away
    if pd.isna(text_value):
        return None
    
    try:
        # Convert to string just in case it's mixed
        text_str = str(text_value)
        
        # 2. Find all numbers (including decimals) in the text row
        # This will find ['10', '20'] out of '10-20 kg'
        numbers = re.findall(r'[-+]?\d*\.\d+|\d+', text_str)
        
        # 3. If we didn't find any numbers, return None
        if not numbers:
            return None
        
        # 4. Convert our found string numbers into actual float numbers
        float_numbers = [float(n) for n in numbers]
        
        # 5. Calculate the average of the found numbers (handles single numbers AND ranges!)
        return sum(float_numbers) / len(float_numbers)
        
    except Exception:
        # If anything weird happens, don't crash, just skip the row
        return None
```

```python
# Clean the columns using your new function
df['Clean_Lifespan'] = df['Lifespan (years)'].apply(extract_numeric_average)
df['Clean_Gestation'] = df['Gestation Period (days)'].apply(extract_numeric_average)
df['Clean_Offspring'] = df['Offspring per Birth'].apply(extract_numeric_average)

# Check the first 5 rows to see the old columns next to your beautiful new clean ones!
df[['Lifespan (years)', 'Clean_Lifespan', 'Offspring per Birth', 'Clean_Offspring']].head()
```

```python
df['Clean_Height'] = df['Height (cm)'].apply(extract_numeric_average)
df['Clean_Weight'] = df['Weight (kg)'].apply(extract_numeric_average)
df['Clean_AvgSpeed'] = df['Average Speed (km/h)'].apply(extract_numeric_average)
df['Clean_TopSpeed'] = df['Top Speed (km/h)'].apply(extract_numeric_average)

# Let's run df.info() one more time to see our beautiful new numeric columns added!
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1. Filter out the extinct categories to keep it focused on active conservation
living_animals = df[~df['Conservation Status'].str.contains('Extinct', na=False, case=False)]

# 2. Filter out extreme rows with missing or 'Varies' labels to keep the graph sharp
clean_status_df = living_animals[~living_animals['Conservation Status'].isin(['Varies', 'Not Applicable'])]

plt.figure(figsize=(11, 5))

# 3. Create the horizontal bar chart using MEDIAN as the estimator
sns.barplot(
    data=clean_status_df, 
    x='Clean_Offspring', 
    y='Conservation Status', 
    estimator=np.median,        # This ensures the bars represent the true typical value (1.0 vs 2.75)
    errorbar=None,              # Removes the messy uncertainty lines
    palette='Spectral',
    order=['Least Concern', 'Near Threatened', 'Vulnerable', 'Endangered', 'Critically Endangered']
)

# 4. Make it look professional with clean titles and labels
plt.title('Typical Number of Offspring per Birth by Conservation Status', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Median Offspring per Birth', fontsize=12)
plt.ylabel('Conservation Status', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.4)

# 5. Clean layout and display
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Filter out extreme outliers and focus on the main statuses
target_statuses = ['Least Concern', 'Endangered', 'Critically Endangered']
filtered_df = df[df['Conservation Status'].isin(target_statuses)]
filtered_df = filtered_df[filtered_df['Clean_Offspring'] <= 20]

# 2. Use col='Conservation Status' to split the graph into side-by-side panels
g = sns.lmplot(
    data=filtered_df,
    x='Clean_Gestation',
    y='Clean_Offspring',
    col='Conservation Status',      # This splits the data into 3 clean, side-by-side graphs
    hue='Conservation Status',      # Keeps the distinct branding colors
    palette='Set1',
    col_order=['Least Concern', 'Endangered', 'Critically Endangered'],
    height=5,
    aspect=0.9,
    sharex=True,                    # Ensures the X-axes match perfectly for easy comparison
    sharey=True,                    # Ensures the Y-axes match perfectly
    scatter_kws={'alpha': 0.5, 's': 60},
    line_kws={'linewidth': 2}
)

# 3. Clean up the titles and labels so they look professional
g.set_titles("{col_name}", weight='bold', size=12)
g.set_axis_labels("Gestation Period (Days)", "Offspring per Birth")

# Add an overarching main title above the subplots
plt.subplots_adjust(top=0.8)
g.fig.suptitle("Reproductive Profiles: Gestation vs. Offspring by Threat Level", fontsize=14, fontweight='bold')

plt.show()
```
